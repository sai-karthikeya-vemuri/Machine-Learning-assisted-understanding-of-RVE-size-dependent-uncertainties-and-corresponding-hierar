```python
import damask
import numpy as np
import matplotlib.pyplot as plt
```

```python
result_files_40=["40Geom_RVE_1_tensionX_273.hdf5","40Geom_RVE_2_tensionX_273.hdf5","40Geom_RVE_3_tensionX_273.hdf5","40Geom_RVE_4_tensionX_273.hdf5","40Geom_RVE_5_tensionX_273.hdf5","40Geom_RVE_6_tensionX_273.hdf5","40Geom_RVE_7_tensionX_273.hdf5","40Geom_RVE_8_tensionX_273.hdf5","40Geom_RVE_9_tensionX_273.hdf5","40Geom_RVE_10_tensionX_273.hdf5"]
result_files_60=["60Geom_RVE_1_tensionX_273.hdf5","60Geom_RVE_2_tensionX_273.hdf5","60Geom_RVE_3_tensionX_273.hdf5","60Geom_RVE_4_tensionX_273.hdf5","60Geom_RVE_5_tensionX_273.hdf5","60Geom_RVE_6_tensionX_273.hdf5","60Geom_RVE_7_tensionX_273.hdf5","60Geom_RVE_8_tensionX_273.hdf5"]
result_files_80=["80Geom_RVE_1_tensionX_273.hdf5","80Geom_RVE_2_tensionX_273.hdf5","80Geom_RVE_3_tensionX_273.hdf5","80Geom_RVE_4_tensionX_273.hdf5","80Geom_RVE_5_tensionX_273.hdf5","80Geom_RVE_6_tensionX_273.hdf5"]
result_files_100=["100Geom_RVE_1_tensionX_273.hdf5","100Geom_RVE_2_tensionX_273.hdf5","100Geom_RVE_3_tensionX_273.hdf5","100Geom_RVE_4_tensionX_273.hdf5","100Geom_RVE_5_tensionX_273.hdf5"]
```

```python
stress_strain_40=[]
stress_strain_60=[]
stress_strain_80=[]
stress_strain_100=[]
```

```python
for result_file in result_files_40:
    result=damask.Result(result_file)
    
    #result.add_stress_Cauchy()
    #result.add_strain()
    #result.add_equivalent_Mises('sigma')
    #result.add_equivalent_Mises('epsilon_V^0.0(F)')
    sigma=[np.average(s[:,0,0]) for s in result.place('P').values()]
    eps=[np.average(e[:,0,0]) for e in result.place('F').values()]
    stress_strain_40.append(np.array([np.array(sigma),np.array(eps)]))
```

```python
for result_file in result_files_60:
    result=damask.Result(result_file)
    
    #result.add_stress_Cauchy()
    #result.add_strain()
    #result.add_equivalent_Mises('sigma')
    #result.add_equivalent_Mises('epsilon_V^0.0(F)')
    sigma=[np.average(s[:,0,0]) for s in result.place('P').values()]
    eps=[np.average(e[:,0,0]) for e in result.place('F').values()]
    stress_strain_60.append(np.array([np.array(sigma),np.array(eps)]))
```

```python
for result_file in result_files_80:
    result=damask.Result(result_file)
    
    #result.add_stress_Cauchy()
    #result.add_strain()
    #result.add_equivalent_Mises('sigma')
    #result.add_equivalent_Mises('epsilon_V^0.0(F)')
    sigma=[np.average(s[:,0,0]) for s in result.place('P').values()]
    eps=[np.average(e[:,0,0]) for e in result.place('F').values()]
    stress_strain_80.append(np.array([np.array(sigma),np.array(eps)]))
```

```python

for result_file in result_files_100:
    result=damask.Result(result_file)
    
    #result.add_stress_Cauchy()
    #result.add_strain()
    #result.add_equivalent_Mises('sigma')
    #result.add_equivalent_Mises('epsilon_V^0.0(F)')
    sigma=[np.average(s[:,0,0]) for s in result.place('P').values()]
    eps=[np.average(e[:,0,0]) for e in result.place('F').values()]
    stress_strain_100.append(np.array([np.array(sigma),np.array(eps)]))
```

```python
result_files_120=["120Geom_RVE_1_tensionX_273.hdf5","120Geom_RVE_2_tensionX_273.hdf5","120Geom_RVE_3_tensionX_273.hdf5"]
stress_strain_120=[]
for result_file in result_files_120:
    result=damask.Result(result_file)
    
    #result.add_stress_Cauchy()
    #result.add_strain()
    #result.add_equivalent_Mises('sigma')
    #result.add_equivalent_Mises('epsilon_V^0.0(F)')
    sigma=[np.average(s[:,0,0]) for s in result.place('P').values()]
    eps=[np.average(e[:,0,0]) for e in result.place('F').values()]
    stress_strain_120.append(np.array([np.array(sigma),np.array(eps)]))
```

```python
labels_40=["40_1","40_2","40_3","40_4","40_5","40_6","40_7","40_8","40_9","40_10"]
labels_60=["60_1","60_2","60_3","60_4","60_5","60_6","60_7","60_8"]
labels_80=["80_1","80_2","80_3","80_4","80_5","80_6"]
labels_100=["100_1","100_2","100_3","100_4","100_5"]
labels_120=["120_1","120_2","120_3"]
fig,ax=plt.subplots(1,5,figsize=(30,5))
for i,j in zip(stress_strain_40,range(0,10)):
    if j>-1:
        ax[0].plot(i[1],i[0]/1000000,label=labels_40[j])
        ax[0].legend()

for i,j in zip(stress_strain_60,range(0,8)):
    if j>-1:
        ax[1].plot(i[1],i[0]/1000000,label=labels_60[j])
        ax[1].legend()

for i,j in zip(stress_strain_80,range(0,8)):
    if j>-1:
        ax[2].plot(i[1],i[0]/1000000,label=labels_80[j])
        ax[2].legend()
for i,j in zip(stress_strain_100,range(0,8)):
    if j>-1:
        ax[3].plot(i[1],i[0]/1000000,label=labels_100[j])
        ax[3].legend()
for i,j in zip(stress_strain_120,range(0,8)):
    if j>-1:
        ax[4].plot(i[1],i[0]/1000000,label=labels_120[j])
        ax[4].legend()

plt.setp(ax,ylim=(220,420))
fig.supxlabel("Deformation Gradient F")
fig.supylabel("Piola Kirchoff Stress P (MPa)")
fig.suptitle("P-F Curves for Thermo-Elasto-Plasticity")
fig.savefig("ThermoElastoPlas_zoomed.png",dpi=300,bbox_inches='tight',transparent=True)
plt.legend()
```
